```python
import cv2
import os
import pandas as pd 

import shutil
from _datetime import datetime
```

```python
cascade_face = cv2.CascadeClassifier('xml/haarcascade_frontalface_default.xml')
cascade_eye = cv2.CascadeClassifier('xml/haarcascade_eye.xml')
cascade_righteye = cv2.CascadeClassifier('xml/haarcascade_righteye_2splits.xml')
cascade_lefteye = cv2.CascadeClassifier('xml/haarcascade_lefteye_2splits.xml')
cascade_smile = cv2.CascadeClassifier('xml/haarcascade_smile.xml')
```

```python
def detection(grayscale, img):
    status = {
        'eyesOpen': False,
        'smile': False,
#         'leftEyeOpen': False,
#         'rightEyeOpen': False
    }
    face = cascade_face.detectMultiScale(grayscale, 1.3, 5)
    for (x_face, y_face, w_face, h_face) in face:
        cv2.rectangle(img, (x_face, y_face),
                      (x_face + w_face, y_face + h_face), (255, 130, 0), 2)
        ri_grayscale = grayscale[y_face:y_face + h_face, x_face:x_face +
                                 w_face]
        ri_color = img[y_face:y_face + h_face, x_face:x_face + w_face]

        eye = cascade_eye.detectMultiScale(ri_grayscale, 1.2, 18)
        for (x_eye, y_eye, w_eye, h_eye) in eye:
            status['eyesOpen'] = True
            cv2.rectangle(ri_color, (x_eye, y_eye),
                          (x_eye + w_eye, y_eye + h_eye), (0, 180, 60), 2)


# #       left eye
#         lefteye = cascade_lefteye.detectMultiScale(ri_grayscale, 1.2, 18)
#         for (x_eye_L, y_eye_L, w_eye_L, h_eye_L) in lefteye:
#             status['leftEyeOpen'] = True
#             cv2.rectangle(ri_color, (x_eye_L, y_eye_L),
#                           (x_eye_L + w_eye_L, y_eye_L + h_eye_L), (0, 180, 60), 2)
            
#             font = cv2.FONT_HERSHEY_SIMPLEX
#             sizes = str(x_eye_L) + "," + str(y_eye_L) + "," + str(
#                 w_eye_L) + "," + str(h_eye_L)

# #             print(sizes)
# #             cv2.putText(img, sizes, (10, 50), font, 0.4, (200, 255, 155),
# #                             1, cv2.LINE_AA)
            
# #         right eye
#         righteye = cascade_righteye.detectMultiScale(ri_grayscale, 1.2, 18)
#         for (x_eye_R, y_eye_R, w_eye_R, h_eye_R) in righteye:
#             status['rightEyeOpen'] = True
#             cv2.rectangle(ri_color, (x_eye_R, y_eye_R),
#                           (x_eye_R + w_eye_R, y_eye_R + h_eye_R), (0, 180, 60), 2)

        smile = cascade_smile.detectMultiScale(ri_grayscale, 1.7, 20)

        for (x_smile, y_smile, w_smile, h_smile) in smile:

            status['smile'] = True
            cv2.rectangle(ri_color, (x_smile, y_smile),
                          (x_smile + w_smile, y_smile + h_smile),
                          (255, 0, 130), 2)


    return img, status
```

```python
MAX_NUMBEROF_PHOTOES_TO_SAVE = 30
num_photos = 0
captureID = 1

if os.path.isdir('Photoes'):
    shutil.rmtree('Photoes')
    os.mkdir('Photoes')
else :
    os.mkdir('Photoes')
    
```

```python
vc = cv2.VideoCapture(0)

times=[]
smile_ratios=[]

while True:
    _, img = vc.read()
    grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    img, status = detection(grayscale, img)
    cv2.imshow('Smile and eye detector', img)

    k = cv2.waitKey(30) & 0xff
    if k == 27:
        break
    print(status)
    
    if (status['smile'] & status['eyesOpen']):
#& status['leftEyeOpen'] & status['rightEyeOpen']):
#cv2.imwrite('Photoes/capture' + str(captureID) + '.png', img)
        captureID = captureID + 1
        num_photos = num_photos + 1

    if num_photos > MAX_NUMBEROF_PHOTOES_TO_SAVE:
        break
        
# ds={'smile_ratio':smile_ratios,'times':times}
# df=pd.DataFrame(ds)
# df.to_csv('smile_records.csv')

vc.release()
cv2.destroyAllWindows()
del (vc)
```
